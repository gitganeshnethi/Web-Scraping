```
import bs4
import pandas as pd
import numpy as np
import requests
```

```
naukri_url = "https://www.monsterindia.com/data-science-jobs.html"
page = requests.get(naukri_url).text
page
my_obj = bs4.BeautifulSoup(page,'html.parser')
```

```
job_titles = my_obj.findAll('span',{'class':'title_in'})
```

```
job_title = [txt.text for txt in job_titles]
job_title
```

```
#method 1 : Get text directly
```

```
company = my_obj.findAll('div',{'class':'jtxt orange'})
[i.text for i in company]
```

```
# method 2: search for anchor tag (child) and get its title attribute
```

```
company_name= []
company = my_obj.findAll('div',{'class':'jtxt orange'})
for i,company in enumerate(company):
    try:
      title = company.find('a',{'class':'jtxt orange'}).get('title')
      company_name.append(title)
    except AttributeError:
      print("missing value found at",i , "th index of titles",'\n')
      company_name.append(float('nan'))
company_name
```

```
job_location = my_obj.findAll('div',{'class':'jtxt jico ico1'})
len(job_location) #ensuring the length of location
```

```
job_locations = [loc.text for loc in job_location]
job_locations
```

```
jobs_data = pd.DataFrame({
    'TITLE':job_title,
    'COMPANY NAME':company_name,
    'JOB LOCATION':job_locations
})
jobs_data.tail(10)
```

```
from google.colab import files

jobs_data.to_csv('sample.csv',index=False)
files.download('sample.csv')
```
